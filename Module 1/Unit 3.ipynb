Matplotlib

Function : It is used to work with graphs and charts

Importing Command : import matplotlib.pyplot as plt

Sub-Functions : plt.plot (For Plotting the graph)
                plt.figure (For Showing and changing the figure Size)
                plt.show  (For Showing the graph )
                plt.xlabel , plt.ylabel (For labelling the axis of the graph )
                plt.title (For adding a title of graph on top )
                plt.tick_params (For adding parameters to values on axis )
                plt.subplot  (To Plot multiple graphs at same time )
                plt.suptitle (To add super titles to subplotted graphs )
                plt.bar  (To Plot bar graph )
                plt.hist  (To Plot Histogram )
                plt.scatter (To Plot Scatter Graph )
                plt.legend (For Adding Legend to Graph )
                ax.annotate (To Add Annotations to the Graph )

# Graph plot using Matplotlib

```python
#import matplotlib
import matplotlib.pyplot as plt

#x-axis co-ordinates
x= [1,2,3]

#y-axis co-ordinates
y= [4,5,6]

#plotting the graph
plt.plot(x,y)

#Showing figure size
fig= plt.figure()

#showing the graph
plt.show()
```

# Showing colors and markers

```python
import numpy as np
import matplotlib.pyplot as plt

x= np.array([2,3,8,7])
y= np.array([1,4,1,9])

plt.plot(x,y, marker= '+',color='g',linestyle='dotted')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line Plot")
plt.show()
```

```python
import random as ran

students =["Aa","Bb","Cc","Dd","Ee"]

marks=[]
for i in range(0,len(students)):marks.append(ran.randint(0,101))

plt.xlabel("Students")
plt.xticks(rotation = 45)
plt.ylabel("Marks")
plt.title("CLASS RECORDS")
plt.plot(students,marks,'g--')
```

```python
# Ticks are the markers denoting the data points on axes
#importing libraries
from matplotlib import pyplot as plt
import numpy as np

#values of x and y
x=[5,10,15,20,25,30,35,40,45,50]
y=[1,4,3,2,7,6,9,8,10,5]

plt.plot(x,y,'g')
plt.xlabel('x')
plt.ylabel('y')

#here we set the size for ticks, rotation and color values
plt.tick_params(axis="x", labelsize=18 , labelrotation=-60 , labelcolor='blue')
plt.tick_params(axis="y", labelsize=12 , labelrotation=-20 , labelcolor='black')

plt.show()
```

# Adding Annotations to Graph

```python
import matplotlib.pyplot as plt
import numpy as np

#draw plot as fig
fig, ax = plt.subplots()


x = np.linspace(0,20,1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

#text annotation to the plot where it indicate maximum value of the
ax.annotate('local maximum', xy=(6.28,1),xytext=(10,4),arrowprops=dict(facecolor='black', shrink =0.05))

#text annotation to the plot where it indicate minimum value of the
ax.annotate('local minimum', xy=(5* np.pi,-1) ,xytext=(2,-6),arrowprops=dict(arrowstyle ="->", connectionstyle ="angle3,angleA=0,angleB=90"))
```

# Adding Annotate & Legend

```python
import numpy as np
import matplotlib.pyplot as plt
# or you can write 
# from matplotlib import pyplot as plt


x = np.arange(0,10)
y = x^2
z = x^3

plt.plot(x,y)
plt.plot(y,z)
plt.plot(x,z)

plt.xlabel('Time')
plt.ylabel('Distance')
plt.title('Graph')

#Annotation

plt.annotate(xy=[2,1],text="First Entry",rotation = 20)
plt.annotate(xy=[4,6],text="Second Entry")
plt.annotate(xy=[8,10],text="Third Entry")

#Adding Legend

plt.legend(["Race 1","Race 2","Race 3"],loc=4)

plt.show
```

# Pie Plot

Pie chart is a circular statistical plot that can display only one series of data 

the area of the chart is the total % of given data

```python
#import libraries

import numpy as np
import matplotlib.pyplot as plt

#creating the data set
Branch = ["Computer","Electrical","Mechanical","Civil","Automobile","IT"]

data = [50,34,40,29,12,23]
explode =(0.1,0,0,0,0.2,0)
#create chart

fig =plt.figure(figsize=(10,7))
plt.pie(data,labels=Branch, explode=explode, autopct='%1.1f%%',shadow=True)

plt.show()
```

# 3D Plotting

```python
#import matplot3d toolkit , numpy, & matplotlib

import numpy as py
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d #3d toolkit library

fig = plt.figure()

#syntax for 3d projection

ax= plt.axes(projection= '3d')

#defining all 3 axes

z = np.linspace(0,1,100)
x = z*np.sin(25*z)
y = z*np.cos(25*z)

#3d plotting

ax.plot3D(x,y,z, 'brown')
ax.set_title('3D ploting')
plt.show()
```
